```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "7"

from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
from ultralytics.models.yolo.detect.val import DetectionValidator
```

```python
model = YOLO("/home/gayapath/PROJECTS/logs/ATS/YOLOv8-L-READ-1/weights/best.pt")
img_path = "/home/gayapath/PROJECTS/DATASETS/READ_2016/test/Images/Seite0405.JPG"

results = model.predict(img_path, conf=0.5, iou=0.5, verbose=False)
r = results[0]

annotated_bgr = r.plot()
annotated_rgb = cv2.cvtColor(annotated_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(annotated_rgb)
plt.axis("off")
plt.show()
```

```python
weights = "/home/gayapath/PROJECTS/logs/ATS/YOLOv8-L-READ-1/weights/best.pt"
data = "read2016.yaml"

model = YOLO(weights)

ious = [x / 100 for x in range(30, 96, 5)]
confs = [x / 100 for x in range(30, 96, 5)]

results = []
best_mAP, best_iou, best_conf = 0, 0, 0
for iou in ious:
    for conf in confs:
        
        args = dict(
            model=weights,
            data=data,
            split="val",     # or "val"
            imgsz=1280,
            conf=conf,
            iou=iou,
            max_det=2000,
            device=7,
        )
        
        validator = DetectionValidator(args=args)
        validator(model=model.model)          # run validation
        metrics = validator.metrics
        
        mAP = metrics.box.map
        results.append([iou, conf, mAP])

        if mAP > best_mAP:
            best_mAP = mAP
            best_iou = iou
            best_conf = conf

print(f"Best: mAP: {best_mAP}")
print(f"Best: iou: {best_iou}")
print(f"Best: conf: {best_conf}")
```
